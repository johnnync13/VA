```python
from skimage import *

# Histogram RGB
# Look for the difference between histograms with imagen It - It-1 [val absoluto] Media y diferencias
# Plot de las diferencias

def histogram_n_bins(image, bins):
    
    image_float = img_as_float(image) # Conversion image's dtype to float 
    
    red_channel, bin_centers = exposure.histogram(image_float[:,:,0], nbins = bins)
    green_channel, bin_centers = exposure.histogram(image_float[:,:,1], nbins = bins)
    blue_channel, bin_centers = exposure.histogram(image_float[:,:,2], nbins = bins)
    
    return (red_channel, green_channel, blue_channel)
```

```python
import numpy as np
import skimage
from skimage import io
import matplotlib.pyplot as plt

#method to calculate the mean
def calcul_mitjana():
    ic = io.ImageCollection('images/Barcelona-sequence/*.png')

    red_init, green_init, blue_init = histogram_n_bins(ic[0],8)

    plot_data = []

    for fotograma in ic[1:]:
        red, green, blue = histogram_n_bins(fotograma,8)

        red_diff = np.linalg.norm(red-red_init)
        green_diff = np.linalg.norm(green-green_init)
        blue_diff = np.linalg.norm(blue-blue_init)

        mean = (red_diff+green_diff+blue_diff)/(3*8) # mean of the 3 channels by 8 bins

        plot_data.append(mean) # add mean value
        
        red_init = red
        green_init = green
        blue_init = blue
        
    plot_data[0] = 0
    return plot_data
```

- Which measure can be used in order to visually distinguish the shots in a plot? Explain your solution.
    
    We distinguish the jump of frame by evaluating **the mean value of RGB** in the histogram representation.

```python
import numpy as np
import skimage
%matplotlib inline

def find_scenes():

    hist_collection = list()
    ph = [0]
    pos = 0

    for i in ic:
        hist = histogram_n_bins(i, 8)
        if len(hist_collection)!=0:
            red_difference = abs(hist[0] - hist_collection[len(hist_collection) - 1][0])
            green_difference = abs(hist[1] - hist_collection[len(hist_collection) - 1][1])
            blue_difference = abs(hist[2] - hist_collection[len(hist_collection) - 1][2])

            rgb_difference = [red_difference, green_difference, blue_difference]
            rgb_mean = int(np.mean(rgb_difference))

            if rgb_mean > 10000:
                ph.append(pos-1)
                ph.append(pos)

        hist_collection.append((hist[0], hist[1], hist[2]))
        pos+=1
    
    ph.append(pos-1)
    return ph

def plot_images(images, titles, rows, cols, fig_size):
    
    figuras, fig_channels = plt.subplots(ncols=cols, nrows=rows,figsize=fig_size)
    idx = 0
    for i in range(rows):
        for j in range(cols):
            fig_channels[i][j].imshow(ic[images[idx]])
            fig_channels[i][j].set_title(titles[idx])
            idx += 1
    plt.show()
    
def generate_titles(n):

    init_titles = ["Initial image shot: " + str(i+1) for i in range(n)]
    final_titles = ["Final image shot: " + str(i+1) for i in range(n)]

    titles = [None]*(len(init_titles)+len(final_titles))
    titles[::2] = init_titles
    titles[1::2] = final_titles
    return titles

```

```python
images = find_scenes()
titles = generate_titles(4) 
plot_images(images, titles, 4, 2, (15,15))
```

Apply the background substraction algorithm (check theory material).

Visualize, for each shot of the video:

**Hint**: You can construct a mask obtained from the original image and the background in order to know which parts of the image form part from the foreground and recover from the original image just the foreground regions.

```python
%matplotlib inline

def plot_images(images, titles, rows, cols, fig_size):
   
    representation, position = plt.subplots(ncols=cols, nrows=rows,figsize=fig_size)
    idx = 0
    for i in range(rows):
        for j in range(cols):
            position[i][j].imshow(images[idx])
            position[i][j].set_title(titles[idx])
            idx += 1
    plt.show()
    
def generate_titles2():
    init_titles = ["Background Image scene " + str(i+1) for i in range(4)]
    final_titles = ["Foreground Image Scene " + str(i+1) for i in range(4)]

    titles = [None]*(len(init_titles)+len(final_titles))
    titles[::2] = init_titles
    titles[1::2] = final_titles
    return titles
    
def group_scene_boundaries(l, n):
    return list(zip(*[l[i::n] for i in range(n)]))

def generate_bg_fg(inicial, final):
    size = list(ic[0].shape)

    background = np.zeros((size))

    images = list()
    red_channel = list()
    green_channel = list()
    blue_channel = list()
    #images.append(img_as_float(ic[:]))
    #Obtain the foreground image (e.g. using the median filter).
    
    for i in range(inicial, final + 1):
        
        images.append(img_as_float(ic[i]))
        red_channel.append(img_as_float(ic[i][:,:,0])) #add to a list each channel of each image
        green_channel.append(img_as_float(ic[i][:,:,1]))
        blue_channel.append(img_as_float(ic[i][:,:,2]))
        
    background[:,:,0] = np.median(red_channel,axis=0) #apply median filter for each channel
    background[:,:,1] = np.median(green_channel,axis=0)
    background[:,:,2] = np.median(blue_channel,axis=0)

    foreground = np.abs(images[0] - background) #foreground is the difference between the original image and the background
    
    return (background, foreground)

scenes = group_scene_boundaries(find_scenes(), 2) # obtain the shots of the video (change of scenes)

backgrounds = list()
for i in range(len(scenes)):
    init = scenes[i][0]
    fin = scenes[i][1]
    bg = generate_bg_fg(init,fin) #generate its background and foreground
    backgrounds.append(bg[0])
    backgrounds.append(bg[1])

plot_images(backgrounds, generate_titles2(), 4, 2, (15,15))
```

The n_segments represents the approximate number of labels of the segmented image. The compactness balances the color proximity, with the higher values, it makes superpixel shapes more cubic as we can see in the first image. This method is based in the k-means clustering algorithm. It consists in selecting random points as initial clusters centers, determinating points in this center and using the Euclidian distance to find the closest point. This process will be repeated until converging to a local minimum.
